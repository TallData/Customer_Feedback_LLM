```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load preprocessed data
file_path = 'data/preprocessed_data.csv'  # Adjust the file path accordingly
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()

bins = [-1, -0.8, -0.5, -0.2, 0.2, 0.5, 0.8, 1]  # Define 9 bins for sentiment polarity
labels = [ 'Negative', 'Moderately Negative', 'Slightly Negative', 'Neutral', 'Slightly Positive', 'Moderately Positive', 'Positive'] 
df['sentiment_category'] = pd.cut(df['sentiment_polarity'], bins=bins, labels=labels, include_lowest=True)

# Visualize the distribution using a bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment_category', data=df, palette='viridis')
plt.title('Distribution of Sentiment Labels')
plt.xlabel('Sentiment Categories')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
